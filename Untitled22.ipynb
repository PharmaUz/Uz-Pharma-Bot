```python
import pandas as pd
import requests

# --- 1ï¸âƒ£ Fetch PharmAgency API ---
pharmagency_url = "https://api.pharmagency.uz/drug-catalog-api/v2/referent-price/all"
headers = {"Accept": "application/json", "User-Agent": "Mozilla/5.0"}

try:
    resp = requests.get(pharmagency_url, headers=headers, timeout=30, verify=False)
    pharm_json = resp.json()

    # Check if 'result' and 'content' exist
    if "result" in pharm_json and "content" in pharm_json["result"]:
        pharm_df = pd.json_normalize(pharm_json["result"]["content"])
        print("âœ… PharmAgency data loaded with columns:", pharm_df.columns.tolist()[:10], "...")
    else:
        print("âš ï¸ No 'result.content' found â€” check API response structure.")
        pharm_df = pd.DataFrame()

except Exception as e:
    print("âš ï¸ PharmAgency fetch failed:", e)
    pharm_df = pd.DataFrame()

# --- 2ï¸âƒ£ Fetch UzPharm-Control API (reuse if already done) ---
uz_url = "https://www.uzpharm-control.uz/registries/api_mpip/server-response.php?draw=1&start=0&length=5000"

try:
    uz_resp = requests.get(uz_url, timeout=30, verify=False)
    uz_data = uz_resp.json()
    uz_df = pd.json_normalize(uz_data["data"])
    print("âœ… UzPharm-Control columns:", uz_df.columns.tolist()[:10], "...")
except Exception as e:
    print("âš ï¸ UzPharm-Control fetch failed:", e)
    uz_df = pd.DataFrame()

# --- 3ï¸âƒ£ Clean UzPharm-Control data ---
uz_df = uz_df.rename(columns={
    "medicine_name": "brand_name",
    "inn_name": "inn",
    "producer_name": "manufacturer",
    "certificate_number": "registration_code"
})[["brand_name", "inn", "manufacturer", "registration_code"]]

# --- 4ï¸âƒ£ Merge (only if both DataFrames exist) ---
if not pharm_df.empty and "brand_name" in uz_df.columns:
    # Try to find PharmAgencyâ€™s matching column names
    merge_col = None
    for c in pharm_df.columns:
        if "name" in c.lower():
            merge_col = c
            break

    print(f"ğŸ”— Merging on UzPharm.brand_name â†” PharmAgency.{merge_col}")
    merged = pd.merge(uz_df, pharm_df, how="outer", left_on="brand_name", right_on=merge_col)
else:
    merged = uz_df.copy()

print("âœ… Final merged data shape:", merged.shape)
merged.head()
```

```python
# --- STEP 7: Clean merged dataset & create fraud-detection flags ---

import numpy as np

df = merged.copy()

# --- 1ï¸âƒ£ Normalize text columns ---
text_cols = ['brand_name', 'inn', 'manufacturer_x', 'registration_code', 'innName', 'manufacturer_y']
for c in text_cols:
    if c in df.columns:
        df[c] = df[c].astype(str).str.strip().str.lower()

# --- 2ï¸âƒ£ Handle price columns (from PharmAgency) ---
price_cols = ['price', 'wholesalePrice', 'retailPrice']
for c in price_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

# --- 3ï¸âƒ£ Create fraud / quality flags ---
df["missing_price_flag"] = df[["price", "wholesalePrice", "retailPrice"]].isna().all(axis=1)
df["missing_registration_flag"] = df["registration_code"].isna() | (df["registration_code"] == "")
df["missing_brand_flag"] = df["brand_name"].isna() | (df["brand_name"] == "")
df["missing_inn_flag"] = df["inn"].isna() | (df["inn"] == "")
df["mismatch_manufacturer_flag"] = (df["manufacturer_x"] != df["manufacturer_y"]) & df["manufacturer_y"].notna()

# --- 4ï¸âƒ£ Combine fraud flags ---
fraud_columns = [
    "missing_price_flag", "missing_registration_flag",
    "missing_brand_flag", "missing_inn_flag", "mismatch_manufacturer_flag"
]
df["fraud_flag"] = df[fraud_columns].any(axis=1).astype(int)

# --- 5ï¸âƒ£ Save cleaned dataset ---
df.to_excel("ai_pharma_api_cleaned_with_fraud.xlsx", index=False)

print("âœ… Data cleaned and fraud flags created.")
print(f"ğŸ”¢ Final dataset shape: {df.shape}")
print("\nğŸ“Š Fraud flag summary:")
print(df['fraud_flag'].value_counts())
```

```python
import pandas as pd

# --- STEP 8: Dashboard Summary ---

print("âœ… Dataset shape:", merged.shape)
print("âœ… Columns:", merged.columns.tolist())

# Basic summaries
summary = {
    "Total Records": len(merged),
    "Unique INNs": merged["inn"].nunique() if "inn" in merged.columns else 0,
    "Unique Manufacturers": merged["manufacturer_x"].nunique() if "manufacturer_x" in merged.columns else 0,
    "Unique Registration Codes": merged["registration_code"].nunique() if "registration_code" in merged.columns else 0,
    "Missing INN (%)": merged["inn"].isna().mean() * 100 if "inn" in merged.columns else 0,
    "Missing Manufacturer (%)": merged["manufacturer_x"].isna().mean() * 100 if "manufacturer_x" in merged.columns else 0,
    "Missing Registration (%)": merged["registration_code"].isna().mean() * 100 if "registration_code" in merged.columns else 0,
}

summary_df = pd.DataFrame(summary.items(), columns=["Metric", "Value"])

# Fraud flag summary (if exists)
if "fraud_flag" in merged.columns:
    flag_counts = merged["fraud_flag"].value_counts().reset_index()
    flag_counts.columns = ["fraud_flag", "count"]
    print("\nğŸ“Š Fraud Flag Distribution:\n", flag_counts)
else:
    print("âš ï¸ No fraud_flag column found.")

print("\nâœ… Dashboard summary ready!")
summary_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- STEP 10: Visualizations ---

plt.figure(figsize=(6,4))
if "fraud_flag" in merged.columns:
    merged["fraud_flag"].value_counts().plot(kind="bar")
    plt.title("Fraud Flag Distribution")
    plt.xlabel("Fraud Flag")
    plt.ylabel("Count")
    plt.show()
else:
    print("âš ï¸ No fraud_flag column found for visualization.")

# Manufacturer top-10
if "manufacturer_x" in merged.columns:
    top_manu = merged["manufacturer_x"].value_counts().head(10)
    plt.figure(figsize=(8,4))
    sns.barplot(x=top_manu.values, y=top_manu.index)
    plt.title("Top 10 Manufacturers in Dataset")
    plt.xlabel("Count")
    plt.ylabel("Manufacturer")
    plt.show()

# INN top-10
if "inn" in merged.columns:
    top_inn = merged["inn"].value_counts().head(10)
    plt.figure(figsize=(8,4))
    sns.barplot(x=top_inn.values, y=top_inn.index)
    plt.title("Top 10 INNs in Dataset")
    plt.xlabel("Count")
    plt.ylabel("INN")
    plt.show()

print("âœ… Step 10 completed â€” Dashboard and visualizations ready.")
```

```python
# --- Step 11ï¸âƒ£: Export Final Dataset and Summary Report ---

# 1ï¸âƒ£ Save cleaned merged data
merged.to_excel("merged_pharma_dataset.xlsx", index=False)
print("âœ… Cleaned dataset exported as 'merged_pharma_dataset.xlsx'")

# 2ï¸âƒ£ Create summary report DataFrame (if not already)
summary = pd.DataFrame({
    "Metric": ["Total Records", "Unique INNs", "Unique Manufacturers", "Unique Registration Codes",
                "Missing INN (%)", "Missing Manufacturer (%)", "Missing Registration (%)"],
    "Value": [merged.shape[0],
              merged["inn"].nunique(),
              merged["manufacturer_x"].nunique() if "manufacturer_x" in merged.columns else merged["manufacturer"].nunique(),
              merged["registration_code"].nunique(),
              merged["inn"].isna().mean(),
              merged["manufacturer_x"].isna().mean() if "manufacturer_x" in merged.columns else merged["manufacturer"].isna().mean(),
              merged["registration_code"].isna().mean()]
})

# 3ï¸âƒ£ Export dashboard summary
summary.to_excel("dashboard_summary.xlsx", index=False)
print("âœ… Dashboard summary exported as 'dashboard_summary.xlsx'")

# 4ï¸âƒ£ Optional: merge both into a single Excel workbook
with pd.ExcelWriter("pharma_final_report.xlsx") as writer:
    merged.to_excel(writer, sheet_name="Merged_Data", index=False)
    summary.to_excel(writer, sheet_name="Summary", index=False)
print("ğŸ“Š Full report saved as 'pharma_final_report.xlsx'")
```

```python
import pandas as pd

print("ğŸš€ Fetching API data...")

# --- 1ï¸âƒ£ Fetch PharmAgency API ---
import requests
pharmagency_url = "https://api.pharmagency.uz/drug-catalog-api/v2/referent-price/all"
headers = {"Accept": "application/json", "User-Agent": "Mozilla/5.0"}

try:
    resp = requests.get(pharmagency_url, headers=headers, timeout=30, verify=False)
    pharm_json = resp.json()
    if "result" in pharm_json and "content" in pharm_json["result"]:
        pharm_df = pd.json_normalize(pharm_json["result"]["content"])
        print(f"âœ… PharmAgency data loaded: {pharm_df.shape}")
    else:
        print("âš ï¸ No valid 'result.content' found in PharmAgency API.")
        pharm_df = pd.DataFrame()
except Exception as e:
    print("âš ï¸ PharmAgency fetch failed:", e)
    pharm_df = pd.DataFrame()


# --- 2ï¸âƒ£ Fetch UzPharm-Control API ---
uz_url = "https://www.uzpharm-control.uz/registries/api_mpip/server-response.php?draw=1&start=0&length=5000"

try:
    uz_resp = requests.get(uz_url, timeout=30, verify=False)
    uz_data = uz_resp.json()
    uz_df = pd.json_normalize(uz_data["data"])
    print(f"âœ… UzPharm-Control data loaded: {uz_df.shape}")
except Exception as e:
    print("âš ï¸ UzPharm-Control fetch failed:", e)
    uz_df = pd.DataFrame()


# --- 3ï¸âƒ£ Clean UzPharm-Control data ---
if not uz_df.empty:
    uz_df = uz_df.rename(columns={
        "medicine_name": "brand_name",
        "inn_name": "inn",
        "producer_name": "manufacturer",
        "certificate_number": "registration_code"
    })[["brand_name", "inn", "manufacturer", "registration_code"]]
else:
    print("âš ï¸ UzPharm-Control data empty.")
    uz_df = pd.DataFrame(columns=["brand_name", "inn", "manufacturer", "registration_code"])


# --- 4ï¸âƒ£ Merge both sources ---
if not pharm_df.empty and "brand_name" in uz_df.columns:
    merge_col = None
    for c in pharm_df.columns:
        if "name" in c.lower():
            merge_col = c
            break
    print(f"ğŸ”— Merging on UzPharm.brand_name â†” PharmAgency.{merge_col}")
    merged = pd.merge(uz_df, pharm_df, how="outer", left_on="brand_name", right_on=merge_col)
else:
    merged = uz_df.copy()

print(f"âœ… Final merged dataset shape: {merged.shape}")

# --- 5ï¸âƒ£ Add Fraud Flag ---
merged["fraud_flag"] = 1  # default mock flag, real rules can be applied later
print("âœ… Fraud flag created.")


# --- 6ï¸âƒ£ Handle Manufacturer Column Dynamically ---
manufacturer_col = "manufacturer_x" if "manufacturer_x" in merged.columns else (
    "manufacturer" if "manufacturer" in merged.columns else "manufacturer_y"
)

# --- 7ï¸âƒ£ Generate Dashboard Summary ---
summary = pd.DataFrame({
    "Metric": [
        "Total Records",
        "Unique INNs",
        "Unique Manufacturers",
        "Unique Registration Codes",
        "Missing INN (%)",
        "Missing Manufacturer (%)",
        "Missing Registration (%)"
    ],
    "Value": [
        merged.shape[0],
        merged["inn"].nunique() if "inn" in merged.columns else 0,
        merged[manufacturer_col].nunique() if manufacturer_col in merged.columns else 0,
        merged["registration_code"].nunique() if "registration_code" in merged.columns else 0,
        merged["inn"].isna().mean() if "inn" in merged.columns else 0,
        merged[manufacturer_col].isna().mean() if manufacturer_col in merged.columns else 0,
        merged["registration_code"].isna().mean() if "registration_code" in merged.columns else 0
    ]
})

print("\nâœ… Dashboard summary ready!\n")
print(summary)
```
